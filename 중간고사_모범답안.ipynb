```python
f=open('../CCDS_nucleotide.current.fna')
```

```python
key=fr.split('>')[1].split('\n')[0]
```

```python
value=''.join(fr.split('>')[1].split('\n')[1:])
```

```python
dicName2CDS={}
```

```python
for einfo in fr.split('>')[1:]:
    key=einfo.split('\n')[0]
    value=''.join(einfo.split('\n')[1:])
    dicName2CDS[key]=value
```

```python
dicName2CDS
```

```python
dicName2Protein={}
```

```python
for key in dicName2CDS :
    cds=dicName2CDS[key]
    protein=''
    for x in range(int(len(cds)/3)):
        protein+=gencode[cds[3*x:3*x+3].upper()]
    dicName2Protein[key]=protein
```

위 문제에서 만든 dicName2CDS, dicName2Protein 을 활용하여 header가 유전자이름, CDS서열, Protein서열, CDS길이, Protein길이 로 구성되어있는 pandas dataframe을 만들어보자

```python
dic={'genename':[],
    'cds':[],
    'protein':[],
    'length(bp)':[],
    'length(aa)':[]}
```

```python
for gn in dicName2CDS:
    dic['genename'].append(gn)
    dic['cds'].append(dicName2CDS[gn])
    dic['protein'].append(dicName2Protein[gn])
    dic['length(bp)'].append(len(dicName2CDS[gn]))
    dic['length(aa)'].append(len(dicName2Protein[gn]))
```
